**There are a few outliers in the following dataset**

```python
for column in file.iloc[:,0:4].columns:
  q1=file[column].quantile(0.25)
  q3=file[column].quantile(0.75)
  iqr=q3-q1
  ub=q3+1.5*iqr
  lb=q1-1.5*iqr
  file=file[(file[column]<ub) & (file[column]>lb)]
```

**Dependent and independent variables**

```python
dep_var=file.iloc[:,0:1]
indep_var=file.iloc[:,3:4]
```

#Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(indep_var,dep_var,test_size=0.25,random_state=0)
```

```python
print("shape of x_train=",x_train.shape)
print("shape of y_train=",y_train.shape)
```

**Regression Model**

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred.shape
```

**Trial**

```python
trial=np.array([180.0,150.0,130.0,135.0])
trial=trial.reshape(-1,1)
model.predict(trial)
```

```python
print("Coefficient = ",model.intercept_)
print("Slope = ",model.coef_)
```
