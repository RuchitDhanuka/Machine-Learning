```python
numerical_columns = ['kilometers', 'year', 'vehicle_age_years', 'seating_capacity', 'Maximum Horse Power']
```

```python
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    print("Outliers")
    print(column,":",IQR)
```

```python
df.count()
```

```python
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))

    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()

    df = df[~outliers]
```

```python
from wordcloud import WordCloud
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
```

```python
brand_counts = data_encoded[['brand_0', 'brand_1', 'brand_2', 'brand_3', 'brand_4', 'brand_5', 'brand_6']].sum()
plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar')
plt.title('Brand Distribution')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
fuel_type_counts = data_encoded[['fuel_type_0', 'fuel_type_1', 'fuel_type_2']].sum()
plt.figure(figsize=(8, 8))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Fuel Type Distribution')
plt.show()
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(data_encoded['exterior_color']))
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Exterior Colors')
plt.show()
```

```python
transmission_type_counts = data_encoded[['transmission_type_0', 'transmission_type_1']].sum()
plt.figure(figsize=(8, 8))
plt.pie(transmission_type_counts, labels=transmission_type_counts.index, autopct='%1.1f%%', startangle=90)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Transmission Type Distribution')
plt.show()
```

```python
correlation_matrix = data_encoded[numerical_columns].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap (Numeric Data)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

brand_color_counts = {}
for brand_col in ['brand_0', 'brand_1', 'brand_2', 'brand_3', 'brand_4', 'brand_5']:
    brand_color_counts[brand_col] = data_encoded[data_encoded[brand_col] == 1].groupby('exterior_color').size()

plt.figure(figsize=(12, 8))
for brand_col, color_counts in brand_color_counts.items():
    plt.barh(color_counts.index, color_counts.values, label=f'Brand {brand_col[-1]}', alpha=0.7)

plt.title('Stacked Bar Chart: Exterior Colors vs. Encoded Brands')
plt.xlabel('Count')
plt.ylabel('Exterior Colors')
plt.legend(title='Encoded Brands', loc='upper right')
plt.show()
```
