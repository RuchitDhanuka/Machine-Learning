```python
centuries = df['Centuries'].sum()
half_centuries = df['Half_Centuries'].sum()
labels = ['Centuries', 'Half Centuries']
sizes = [centuries, half_centuries]
plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Centuries vs Half Centuries')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Batting_Average'], bins=20, color='skyblue', edgecolor='black')
plt.title('Batting Average Distribution')
plt.xlabel('Batting Average')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['Runs_Scored', 'Batting_Average', 'Bowling_Average', 'Wickets_Taken']])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Batting_Strike_Rate'],df['Bowling_Strike_Rate'], color='red', alpha=0.5)
plt.title('Batting Strike Rate vs Bowling Strike Rate')
plt.xlabel('Batting Strike Rate')
plt.ylabel('Bowling Strike Rate')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df[['Runs_Scored', 'Batting_Average', 'Bowling_Average', 'Wickets_Taken']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Matches_Batted'], df['Runs_Scored'], marker='o', linestyle='-', color='b')
plt.title('Runs Scored over Matches Batted')
plt.xlabel('Matches Batted')
plt.ylabel('Runs Scored')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Runs_Scored'], kde=True, color='skyblue')
plt.title('Distribution of Runs Scored')
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Batting_Strike_Rate'] = (df['Runs_Scored'] / df['Balls_Faced']) * 100
df['Bowling_Strike_Rate'] = (df['Runs_Conceded'] / df['Balls_Bowled']) * 100
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Centuries', y='Bowling_Average', data=df)
plt.title('Bowling Average by Century and Half Century')
plt.xlabel('Centuries')
plt.ylabel('Bowling Average')
plt.show()
```
